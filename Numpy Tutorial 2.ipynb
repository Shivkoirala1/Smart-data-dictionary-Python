```python
# Difference between round floor and ceil
import numpy as np

# Sample array
arr = np.array([3.2, 4.7, -2.5, -3.8])

print("Original Array:", arr)

# Round to nearest integer
print("Round:", np.round(arr,5))

# Floor (always down)
print("Floor:", np.floor(arr))

# Ceil (always up)
print("Ceil:", np.ceil(arr))
```

```python
# Indexing Slicing (Already Covered in Lecture in depth)
import numpy as np
arr1 = np.arange(10)
print("First element:", arr1[0])
print("Last element:", arr1[-1])

# Slicing
print("First 5 elements:", arr1[:5])
print("Elements from index 3 to 7:", arr1[3:8])

arr2 = np.arange(12).reshape((3,4))
print("\narr2:\n", arr2)

# Indexing
print("Element at row 1, col 2:", arr2[1,2])   # second row, third column

# Slicing rows and columns
print("First row:", arr2[0,:])
print("First column:", arr2[:,0])
print("Submatrix (rows 0-1, cols 1-2):\n", arr2[0:2, 1:3])

arr2 = np.arange(8).reshape((2,2,2))
# 3D Array
arr3 = np.arange(8).reshape((2,2,2))
print("\narr3:\n", arr3)

# Indexing
print("Element at [1,0,1]:", arr3[1,0,1])

# Slicing
print("First block:\n", arr3[0])
print("Second block:\n", arr3[1])
```

```python
# reshape
import numpy as np

# Create a 1D array with 12 elements
arr = np.arange(12)
print("Original 1D Array:", arr)

# Reshape into 2D (3 rows, 4 columns)
arr2 = arr.reshape(3, 4)
print("\nReshaped to 3x4:\n", arr2)

# Reshape into 2D (4 rows, 3 columns)
arr3 = arr.reshape(4, 3)
print("\nReshaped to 4x3:\n", arr3)

# Reshape into 3D (2 blocks, 2 rows, 3 columns)
arr4 = arr.reshape(2, 2, 3)
print("\nReshaped to 3D (2x2x3):\n", arr4)
```

```python
# Stacking - horizonal Stacking
arr1 = np.arange(0,12).reshape((3,4))
arr2 = np.arange(12,24).reshape((3,4))

print("arr1:\n", arr1)
print("arr2:\n", arr2)

# Horizontal stacking
hstacked = np.hstack((arr1, arr2))
print("\nHorizontally Stacked Array:\n", hstacked)# horizontally array lai joddinx 01231213 garera
```

```python
# Stacking - vertical Stacking
arr1 = np.arange(0,12).reshape((3,4))
arr2 = np.arange(12,24).reshape((3,4))

print("arr1:\n", arr1)
print("arr2:\n", arr2)

# Vertical stacking
vstacked = np.vstack((arr1, arr2))
print("\nVertically Stacked Array:\n", vstacked)
```

```python
# Splitting - Vertical Splitting 
arr1 = np.arange(0,16).reshape((4,4))
print(arr1)
[arr3,arr2] = np.vsplit(arr1,2)
print(arr3)
arr2
```

```python
# Horizontal Splitting
arr1 = np.arange(0,16).reshape((4,4))
print(arr1)
[arr2,arr3] = np.hsplit(arr1,2)

print( arr3)
arr2
```

#### 3. Create a 2d array with 1 on the border and 0 inside. 

```python
# Code here
import numpy as np

# Define size of the array
n = 4 # you can change this to any size that you want to create 
# n ko value user sanga magna pani sakiyo
# Create an array of zeros
arr = np.zeros((n, n), dtype=int)

# Set the border values to 1
arr[0, :] = 1        # first row
arr[-1, :] = 1       # last row
arr[:, 0] = 1        # first column
arr[:, -1] = 1       # last column

print( arr)


```
