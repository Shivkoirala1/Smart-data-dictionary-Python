#### 1. Matplotlib – Bar Chart Create a bar chart showing the number of students in 5 different classes: Classes: ['A', 'B', 'C', 'D', 'E'] Students: [25, 30, 22, 28, 35] Hint: Use plt.bar()

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
classes = ['A', 'B', 'C', 'D', 'E']
students = [25, 30, 22, 28, 35]
plt.bar(classes, students)
plt.title("Number of Students in Classes")
plt.xlabel("Classes")
plt.ylabel("Students")
plt.show()
```

#### 2. Seaborn – Bar Plot Create the same bar chart in Seaborn and color the bars blue. Hint: Use sns.barplot()

```python
sns.barplot(x=classes, y=students, color="blue")
plt.title("Number of Students in Classes (Seaborn)")
plt.show()
```

#### 3. Matplotlib – Box Plot  Generate a box plot for exam scores: [55, 67, 78, 88, 92, 45, 66, 70, 89, 95] Highlight the outliers Hint: Use plt.boxplot()

```python
scores = [55, 67, 78, 88, 92, 45, 66, 70, 89, 95]
plt.boxplot(scores, flierprops=dict(marker='o', color='red', markersize=8))
plt.title("Exam Scores (Matplotlib)")
plt.show()
```

#### 4. Seaborn – Box Plot Create a box plot in Seaborn for the same scores and change the color to green.
#### Hint: Use sns.boxplot()

```python
sns.boxplot(y=scores, color="green")
plt.title("Exam Scores (Seaborn)")
plt.show()
```

#### 5. Matplotlib – Scatter Plot Plot hours studied vs marks obtained: Hours: [2, 3, 4, 5, 6, 7]
#### Marks: [40, 50, 55, 65, 70, 80]
#### Hint: Use plt.scatter()

```python
hours = [2, 3, 4, 5, 6, 7]
marks = [40, 50, 55, 65, 70, 80]
plt.scatter(hours, marks)
plt.title("Hours Studied vs Marks Obtained")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.show()
```

#### 6. Seaborn – Scatter Plot with Regression Create a scatter plot for the same data and add a regression line.
#### Hint: Use sns.regplot()


```python

sns.regplot(x=hours, y=marks)
plt.title("Hours Studied vs Marks Obtained (Regression)")
plt.show()
```

#### 7. Matplotlib – Pie Chart Create a pie chart for fruits sold: Fruits: ['Apple', 'Banana', 'Orange', 'Mango'] Quantity: [50, 30, 15, 5]
#### Hint: Use plt.pie()

```python
fruits = ['Apple', 'Banana', 'Orange', 'Mango']
quantity = [50, 30, 15, 5]
plt.pie(quantity, labels=fruits, autopct='%1.1f%%')
plt.title("Fruits Sold")
plt.show()
```

#### 8. Seaborn – Categorical Box Plot Plot salaries by department in Seaborn: Departments: ['HR','IT','Sales','Marketing'] Salaries: [40000, 50000, 45000, 60000]
#### Hint: Use sns.boxplot(x=..., y=...)

```python
departments = ['HR','IT','Sales','Marketing']
salaries = [40000, 50000, 45000, 60000]
sns.boxplot(x=departments, y=salaries)
plt.title("Salaries by Department")
plt.show()
```

#### 9. Seaborn – Count Plot Create a count plot of favorite colors from survey data: Colors: ['Red', 'Blue', 'Green', 'Red', 'Blue', 'Red', 'Green']
#### Hint: Use sns.countplot()

```python
colors = ['Red', 'Blue', 'Green', 'Red', 'Blue', 'Red', 'Green']
sns.countplot(x=colors)
plt.title("Favorite Colors Count")
plt.show()
```
